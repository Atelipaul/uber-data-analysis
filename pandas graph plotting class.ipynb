```python
# matplotlib and seaborn are used for visualization in python

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
df = pd.read_csv("https://raw.githubusercontent.com/waleolajumoke/data/main/dm_office_sales.csv")
df
```

```python
# creating plot

x = np.arange(1,6)
y = x * 2
plt.xlabel("X Axis")    # labelling
plt.ylabel("Y Axis")
plt.title("Sample Graph")
plt.plot(x,y)          #plotting
plt.savefig("my-sample-graph.jpg")  # savingÂ theÂ graph
```

```python
plt.figure(figsize=(10,4))   # resizing
x = np.arange(1,6)
y = x * 2
plt.xlabel("X Axis")    # labelling
plt.ylabel("Y Axis")
plt.title("Sample Graph")
plt.plot(x,y)          #plotting
plt.savefig("my-sample-graph.jpg")  # savingÂ theÂ graph   
```

```python
# stying your graphg
plt.figure(figsize=(10,4))
plt.plot(x,y, color="red", marker="o", markersize=10, label="2020 sales")
plt.plot(x , x**3, color="green", marker="+", label="2021 sales")
plt.legend()
```
